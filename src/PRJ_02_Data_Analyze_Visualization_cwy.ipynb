#### 2.2 데이터 분석 및 시각화
##### 2.2.1 신채발달현황
* 연간 초등학생 학년별 평균 키 변화
    * 최근 연도가 될수록 약간씩 더 성장하고 있다.

```python
df1 = df.groupby(['학년도', '학년']).agg({'키_cm' : 'mean'}).sort_values(['학년도', '학년']).reset_index()
df1.head()
```

```python
years = [ i  for i in np.arange(2015, 2022+1) ]
years.remove(2020)

heights_by_grade = {
    '1학년': df1[df1['학년'] == 1]['키_cm'].to_list(),
    '2학년': df1[df1['학년'] == 2]['키_cm'].to_list(),
    '3학년': df1[df1['학년'] == 3]['키_cm'].to_list(),
    '4학년': df1[df1['학년'] == 4]['키_cm'].to_list(),
    '5학년': df1[df1['학년'] == 5]['키_cm'].to_list(),
    '6학년': df1[df1['학년'] == 6]['키_cm'].to_list(),
}

# 다중 라인 그래프 그리기
plt.figure(figsize=(6,5))

for grade, heights in heights_by_grade.items():
    plt.plot(years, heights, label=grade)

plt.title('학년별 평균 키 변화추이', fontdict=font, pad=15)
plt.xlabel('Year')
plt.ylabel('평균 키 (cm)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.xticks(years)
plt.show()
```

* 첫번째 챠트 :  연간 초등학생 학년별 평균 몸무게 변화
    * COVID-19 팬대믹에(2020년부터) 의한 평균몸무게 급증 (비대면 학습 등의 영향으로 보임)

* 두번째 챠트 :  연간 초등학생 6학년 성별 평균 몸무게 변화
    * 연간 몸무게 변화율이 높았던 6학년 학생들의 표본에서 성별 몸무게 변화 추이 확인
    * 남학생의 2021년도 몸무게 급증현상이 더 두드러짐

```python
df2 = df.groupby(['학년도', '학년']).agg({'몸무게_kg' : 'mean'}).sort_values(['학년도', '학년']).reset_index()
df2.head()
```

```python
df3 = df[df['학년']==6].groupby(['학년도', '성별']).agg({'몸무게_kg' : 'mean'}).sort_values(['학년도',  '성별']).reset_index()
df3.head()
```

```python
years = [ i  for i in np.arange(2015, 2022+1) ]
years.remove(2020)

heights_by_grade = {
    '1학년': df2[df2['학년'] == 1]['몸무게_kg'].to_list(),
    '2학년': df2[df2['학년'] == 2]['몸무게_kg'].to_list(),
    '3학년': df2[df2['학년'] == 3]['몸무게_kg'].to_list(),
    '4학년': df2[df2['학년'] == 4]['몸무게_kg'].to_list(),
    '5학년': df2[df2['학년'] == 5]['몸무게_kg'].to_list(),
    '6학년': df2[df2['학년'] == 6]['몸무게_kg'].to_list(),
}

# 다중 라인 그래프 그리기
fig, ax = plt.subplots(1,2,figsize=(10,5))

for grade, heights in heights_by_grade.items():
    ax[0].plot(years, heights, label=grade)

ax[0].set_title('학년별 평균 몸무게 변화', fontdict=font, pad=55)
ax[0].set_xlabel('Year')
ax[0].set_ylabel('평균 몸무게 (kg)')
ax[0].legend(loc='upper center', bbox_to_anchor=(0.5, 1.20), ncol=3)
ax[0].grid(True)
ax[0].set_xticks(years)


##########################################################3
# 가상의 연도별 성별 평균 몸무게 데이터
male_weights = df3[df3['성별'] == '남']['몸무게_kg'].to_list()
female_weights = df3[df3['성별'] == '여']['몸무게_kg'].to_list()

ax[1].plot(years, male_weights, marker='o', label='남학생(6학년)')
ax[1].plot(years, female_weights, marker='o', label="여학생(6학년)")

ax[1].set_title('연간 성별 평균 몸무게 변화', fontdict=font, pad=55)
ax[1].set_xlabel('연도')
ax[1].set_ylabel('평균 몸무게 (kg)')
ax[1].legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2)
ax[1].grid(True)
ax[1].set_xticks(years)  # x축 눈금을 연도로 설정
fig.subplots_adjust(left=0.1, right=0.9, top=0.85, bottom=0.2, wspace=0.4)
```

* 몸무게 대비 키 비율의 변화 확인
    * 비만여부 데이터는 2018년,2019년,2021년 데이터에 제공되지 않아, 몸무게 대비 키 비율을 확인함
    * IQR 범위(전체 데이터의 25%~75%지점)을 Normal 등급으로 정의
        * Lower 등급 : Q1 보다 낮은 경우
        * Upper 등급 : Q3 보다 높은 경우
    * 최솟값(IQR 1사분위값에서 1.5배 낮은 범위의 값) : box plot 에서 이상치 제거시 사용
    * 최대값(IQR 3사분위값에서 1.5배 높은 범위의 값) : box plot 에서 이상치 제거시 사용

```python
df4 = df
df4['몸무게 대비 키 비율'] = df['몸무게_kg']/df['키_cm'] * 100

# 몸무게 대비 비율로 Lower, Normer, Upper 등급 분류하기
Q1 = df4['몸무게 대비 키 비율'].quantile(0.25)
Q3 = df4['몸무게 대비 키 비율'].quantile(0.75)

# IQR 계산
IQR = Q3 - Q1

upper_fence = Q3 + 1.5 * IQR
lower_fence = Q1 - 1.5 * IQR

print('IQR:' , IQR, 'Q3:', Q3, 'Q1:', Q1, 'Upper기준:', upper_fence,'초과 ', 'Lower기준:', lower_fence, '미만 ')

df4.loc[df4['몸무게 대비 키 비율'] < Q1, '몸무게등급'] = 'Lower'
df4.loc[df4['몸무게 대비 키 비율'] > Q3, '몸무게등급'] = 'Upper'
df4['몸무게등급'] = df4['몸무게등급'].fillna('Normal')
df4[['학년도', '학년', '몸무게_kg', '키_cm', '몸무게 대비 키 비율', '몸무게등급']].head()
```

* 연간 초등학생 몸무게 대비 키 비율 분포 변화
    * 이상치를 제외하고 보았을 경우에도 2019년 이후 몸무게 비중이 높아진것을 확인가능

```python
plt.figure(figsize=(6,5))
plt.title('연간 초등학생 몸무게 대비 키 비율 분포', fontdict=font, pad=15)
df4_filtered = df4[(df4['몸무게 대비 키 비율'] > lower_fence+5) & (df4['몸무게 대비 키 비율'] < upper_fence-5)] # 이상치 없이 box 플롯그리기
sns.boxplot(y='몸무게 대비 키 비율', x='학년도', data=df4_filtered);
```

* 연도별 몸무게 등급별 비율
    * 위에서 정의한 몸무게 등급인 (Lowerr, Normal, Upper)의 분포를 연도별로 확인한다.
    * 몸무게 Upper 등급의 비율이 2021년도 부터 높아진 것을 확인할 수 있다.

```python
df5 = df4.groupby(['학년도', '몸무게등급'])['학년'].count().reset_index().rename(columns={'학년':'학생수'})
df4_sub = df4.groupby('학년도')['학년'].count().reset_index().rename(columns={'학년':'전체학생수'})
df5 = df5.merge(df4_sub, on='학년도')
df5['비율'] = df5['학생수']/df5['전체학생수']
df5.head()
```

```python
# 누적 막대 그래프 그리기
plt.figure(figsize=(6, 5))
sns.barplot(x='학년도', y='비율', hue='몸무게등급', data=df5)
plt.title('연도별 몸무게 등급별 비율' , fontdict=font, pad=15)
plt.xlabel('연도')
plt.ylabel('비율')
plt.legend(title='몸무게 등급')
plt.grid(True, axis='y')  # y 축에 그리드 추가
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
df6 = df4[['몸무게등급', '라면', '음료수','패스트푸드', '육류', '우유_유제품', '과일', '채소(김치제외)', '아침식사', '주3회이상운동', '하루수면량', '하루TV시청2시간이상', '2시간이상게임']]
mapping = {'Lower':1, 'Normal':2, 'Upper':3}
df6.loc[:, '몸무게등급'] = df6['몸무게등급'].map(mapping).astype(int)
```

```python
#몸무게등급, 라면, 음료수, 패스트푸드, 육류, 우유_유제품, 과일, 채소(김치제외), 아침식사, 주3회이상운동, 하루수면량, 하루TV시청2시간이상, 2시간이상게임
plt.figure(figsize=(12, 5))
sns.heatmap(round(df6.corr(),3), cmap='Reds', annot=True)
plt.title('몸무게와의 상관관계', fontdict=font, pad=15)
plt.show()
```

* 정상시력/ 시력저하(나안 좌 또는 우 시력이 0.5이하이거나 교정중)인 학생수의 비율
    * 시력검사를 수행한 학생들에 대해서만 분석(전체 253717 학생중 80399만 시행함)
    * 15%정도만 시력저하 상태임
* 청력 정상유무 비율
    * 청력은 거의 다 정상

```python
df7 = df[(df['시력_교정_좌'].notnull()) | (df['시력_교정_우'].notnull()) | (df['시력_나안_우'].notnull()) | (df['시력_나안_좌'].notnull())][['학년도', '학년', '시력_교정_좌', '시력_교정_우', '시력_나안_좌', '시력_나안_우']]

df7['시력저하유무'] = (df['시력_교정_좌'].notnull()) | (df['시력_교정_우'].notnull()) |  (df['시력_나안_우'] <= 0.5) |   (df['시력_나안_좌'] <= 0.5)
```

```python
df8 = df
df8 = df[(df['청력_좌'].notnull()) | (df['청력_우'].notnull()) | (df['청력_좌'] != '검사안함') | (df['청력_우'] != '검사안함')][['학년도', '학년', '청력_좌', '청력_우']]

df8['청력정상유무'] = (df['청력_좌'] =='이상') | (df['청력_우'] == '이상')
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,4))

# count of col (pie chart)
slices = df7['시력저하유무'].value_counts().values
activities =['시력정상','시력저하']
ax[0].pie(slices, labels=activities, shadow=True, autopct='%1.1f%%')
ax[0].set_title('시력건강', fontdict=font, pad=10)

slices1 = df8['청력정상유무'].value_counts().values
activities1 =['청력정상','청력이상']
ax[1].pie(slices1, labels=activities1, shadow=True, autopct='%1.1f%%')
ax[1].set_title('청력건강', fontdict=font, pad=10)


plt.show()
```
